```python
import seaborn as sns
import matplotlib.pyplot as plt
planets = sns.load_dataset('planets')
planets
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(
    x=planets['mass'],
    y=planets['orbital_period'],
    data=planets
)

plt.show()
```

```python
planets_clean = planets.dropna(subset=['mass', 'orbital_period'])
plt.figure(figsize=(10, 6))
sns.scatterplot(x='mass', y='orbital_period', data=planets_clean)
plt.title('Scatter Plot of Planet Mass vs Orbital Period')
plt.show()
```

```python
plt.hist(planets['mass'], bins=50, edgecolor='black')
plt.title('Distribution of Planet Mass')
plt.show()
```

```python
sns.lineplot(x = "year", y = "orbital_period", data = planets )
plt.xlabel("YEAR")
plt.ylabel("Orbital Period")
plt.show()
```

```python
from bokeh.plotting import show
```

```python
ind_values = np.arange(-5, 5, 1) 
dep_values = ind_values ** 2
p = figure(title='Plot of y = x^2',x_axis_label='Independent Values(x)',y_axis_label='Dependent Values(y)')
p.line(ind_values, dep_values, line_width=2, color='blue', legend_label='y = x^2')
p.circle(ind_values, dep_values, size=8, color='red')
show(p)
```

```python
independent_values = np.arange(-5, 6, 1)
dependent_values = independent_values ** 2
p = figure(title='Line and Scatter Plot of y = x^2',x_axis_label='Independent Values (x)',y_axis_label='Dependent Values (y)')
p.line(independent_values, dependent_values, line_width=2, color='blue', legend_label='Line: y = x^2')
p.circle(independent_values, dependent_values, size=8, color='red', legend_label='Scatter: Points')
p.legend.location = "top_left"
show(p)
```

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
cancer
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
from sklearn.preprocessing import StandardScaler
df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
scaler=StandardScaler()
scaler.fit(df) 
```

```python
scaler = StandardScaler()
scaler.fit(df)
scaled_data = scaler.transform(df)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)  # Reduce to 2 components
pca.fit(scaled_data)
X_pca = pca.transform(scaled_data)
```

```python
print("Original scaled data shape:", scaled_data.shape)
print("PCA reduced data shape:", X_pca.shape)
```

```python
plt.scatter(X_pca[:,0], X_pca[:,1], c = cancer['target'])
plt.xlabel('First Principle Component')
plt.ylabel('Second principal Component')
plt.show()
```
