```python
def sort (nums):
    for i in  range (len (nums)-1, 0,-1):
        for j in range (i):
            if nums [j]>nums[j+1]:
                temp=nums[j]
                nums[j]=nums[j+1]
                nums[j+1]=temp
#ex
nums=[5,3,8,6,7,2]
sort (nums)
print(nums)
```

```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = left + (right - left) // 2
        print(mid)
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
arr=[6,8,10,14,33,34,43,55,57,89]
key=59
result=binary_search(arr,key)
if result !=1:
    print(f"key {key} found at index{result}")
else:
    print("key not found in the array")
```
