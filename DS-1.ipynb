```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (replace 'your_dataset.csv' with the actual filename)
covid_data = pd.read_csv('file3/covid_data.csv')
# Set the Date column as the index
covid_data.set_index('Date', inplace=True)

# Time Series Plot
plt.figure(figsize=(14, 6))
for metric in ['Cases', 'Recoveries', 'Deaths', 'Vaccination', 'Testing', 'Hospitalizations']:
    plt.plot(covid_data.index, covid_data[f'Country_A_{metric}'], label=f'Country A {metric}')
plt.title('Daily Trends for Country A')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
# Bar Plot - Proportion of Cases, Recoveries, and Deaths
cumulative = covid_data.iloc[-1].filter(like='Country_A').groupby(lambda x: x.split('_')[-1]).sum()
cumulative[['Cases', 'Recoveries', 'Deaths']].plot(kind='bar', figsize=(8, 5))
plt.title('Proportion of Cases, Recoveries, and Deaths (Country A)')
plt.ylabel('Count')
plt.show()
```

```python
# Heatmap - Correlation Between Factors
corr_matrix = covid_data.filter(like='Country_A').corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix for Country A')
plt.show()
```

```python
# Scatter Plot - Vaccination Rate vs Hospitalization Rate
plt.figure(figsize=(8, 5))
sns.regplot(x=covid_data['Country_A_Vaccination'], y= covid_data['Country_A_Hospitalizations'])
plt.title('Vaccination Rate vs Hospitalization Rate (Country A)')
plt.xlabel('Vaccination')
plt.ylabel('Hospitalizations')
plt.show()
```

```python
fig, axs = plt.subplots(3, 2, figsize=(12, 12))
metrics = ['Cases', 'Recoveries', 'Deaths', 'Testing', 'Vaccination', 'Hospitalizations']
for i, metric in enumerate(metrics):
    row, col = divmod(i, 2)
    axs[row, col].plot(covid_data.index, covid_data[f'Country_A_{metric}'], label=metric)
    axs[row, col].set_title(f'{metric} (Country A)')
    axs[row, col].legend()
plt.tight_layout()
plt.show()
```

```python
labels = ['Vaccination', 'Hospitalizations']
v = [sum(covid_data['Country_A_Vaccination']), sum(covid_data['Country_A_Hospitalizations'])]  # Replace with appropriate logic

if sum(v) > 0:
    plt.figure(figsize=(6, 6))
    plt.pie(v, labels=labels, autopct='%1.1f%%', colors=['skyblue', 'orange'])
    plt.title('Vaccination vs Hospitalization Distribution')
    plt.show()
else:
    print("Error: The data contains no non-zero values for plotting.")
```
