```python
from numpy.random import random
from bokeh.core.enums import MarkerType
from bokeh.plotting import figure, show

p = figure(title="Bokeh Markers", toolbar_location=None)
p.grid.grid_line_color = None
p.background_fill_color = "#eeeeee"
p.axis.visible = False
p.y_range.flipped = True
N = 10
for i, marker in enumerate(MarkerType):
    x = i % 4
    y = (i // 4) * 4 + 1

    p.scatter(random(N)+2*x, random(N)+y, marker=marker, size=14,
              line_color="navy", fill_color="orange", alpha=0.5)
    p.text(2*x+0.5, y+2.5, text=[marker],
           text_color="firebrick", text_align="center", text_font_size="13px")

show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, LabelSet

# Create a list of marker types
marker_types = ['circle', 'square', 'triangle', 'diamond', 'cross', 'x']
marker_labels = ['Circle', 'Square', 'Triangle', 'Diamond', 'Cross', 'X']

# Create a ColumnDataSource for the markers
source = ColumnDataSource(data=dict(
    x=[1, 2, 3, 4, 5, 6],
    y=[1, 2, 3, 4, 5, 6],
    markers=marker_types,
    labels=marker_labels
))

# Create a figure
p = figure(title="Various Marker Types in Bokeh", x_axis_label='X', y_axis_label='Y', width=600, height=400)

# Add scatter points with different markers
for i in range(len(marker_types)):
    p.scatter('x', 'y', source=source, marker=marker_types[i], size=10, legend_label=marker_labels[i], color="navy", alpha=0.6)

# Add labels for each marker type
labels = LabelSet(x='x', y='y', text='labels', source=source, x_offset=5, y_offset=5)
p.add_layout(labels)

# Show the plot
show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, LabelSet

# Create data for the plot
x = np.arange(1, 7)  # X values
y = np.random.randint(1, 10, size=6)  # Random Y values
marker_types = ['circle', 'square', 'triangle', 'diamond', 'cross', 'x']
marker_labels = ['Circle', 'Square', 'Triangle', 'Diamond', 'Cross', 'X']

# Create a ColumnDataSource for the markers
source = ColumnDataSource(data=dict(
    x=x,
    y=y,
    markers=marker_types,
    labels=marker_labels
))

# Create a figure with no toolbar and a title
p = figure(title="Marker Types with Flipped Y-Axis", toolbar_location=None, width=600, height=400)

# Hide the grid lines
p.grid.grid_line_alpha = 0

# Add scatter points with different markers
for i in range(len(marker_types)):
    p.scatter('x', 'y', source=source, marker=marker_types[i], size=10, legend_label=marker_labels[i], color="navy", alpha=0.6)

# Add labels for each marker type
labels = LabelSet(x='x', y='y', text='labels', source=source, x_offset=5, y_offset=5)
p.add_layout(labels)

# Flip the y-axis so that the highest values are at the bottom
p.y_range.flipped = True

# Show the plot
show(p)
```
