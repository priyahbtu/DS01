```python
from matplotlib import pyplot as plt
import numpy as np 
x=np.arange(1,10,0.1)
y=2*x+5
plt.plot(x,y)
plt.show()
```

```python
years=[1950,1960,1970,1980,1990,2000,2010]
gdp=[300,500,900,1000,1200,1345,1500]
plt.plot(years,gdp,color="green",marker="o",linestyle='solid')
plt.title("nominal gdp")
plt.xlabel("decade")
plt.ylabel("billion of $")
plt.show()
```

```python
x=np.arange(1,10,0.1)
y1=2*x+5
y2=2*x+10
plt.subplot(1,2,1)
plt.plot(x,y1)

plt.subplot(1,2,2)
plt.plot(x,y2)
 
plt.subplot(1,2,2)
```

# two plots in same figure

# bar plot

```python
fruits={'apple':30,'mango':45,'banana':10}
names=list(fruits.keys())
quantity=list(fruits.values())
```

```python
names,quantity
```

```python
plt.bar(names,quantity,color="orange")
plt.title('distribution of fruits')
plt.ylabel('fruits')
plt.xlabel('quantity')
plt.show()
```

# scatter plot

```python
x=[10,20,30,40,50,60,70,80]
a=[1,2,3,4,5,6,7,8]
plt.scatter(x,a,color="pink")
plt.title('random')
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80]
a=[1,2,3,4,5,6,7,8]
b=[5,6,7,2,1,3,2,4]

plt.figure(figsize=(10,10))

plt.scatter(x,a,s=300,marker="3")
plt.scatter(x,b,s=400)
plt.legend(['a','b'])
plt.title('random')
plt.show()
```

```python
movies=['annie hall','ben heir','casablance','gandhi','west side story']
oscar=[5,11,3,8,10]
plt.bar(movies,oscar,color='purple')
plt.title('movies')
plt.ylabel('num_oscar')
plt.xlabel('name of movie')
plt.show()
```

# Pie-chart

```python
fruit=['apple','mango','grapes','oranges']
quantity=[30,40,35,15]
plt.pie(quantity,labels=fruit,autopct='%0.1f%%',shadow=True,explode=(0,0,0,0),colors=['red','pink','yellow','purple'])
plt.show()
```

```python
fruit=['apple','mango','grapes','oranges']
quantity=[30,40,35,15]
pie1=plt.pie(quantity,labels=fruit,autopct='%0.1f%%',radius=2)
pie2=plt.pie([5],colors='white',radius=1)
plt.show()
```
