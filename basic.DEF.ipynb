```python
def Calculate():
    print("palindrome number")
    num = int(input("Enter a number: "))
    if is_palindrome(num):
        print(f"{num} is a palindrome number.")
    else:
        print(f"{num} is not a palindrome number.")
def is_palindrome(num):
    return str(num) == str(num)[::-1]
Calculate()
```

```python
def calculator():
    print("1.addition")
    print("2.substraction")
    print("3.multiplication")
    print("4.division")
    ch=int(input("enter your choice from 1 to 4: "))
    if ch==1:
        n1=int(input("enter number one:"))
        n2=int(input("enter number two:"))
        print(add(n1,n2))
    elif ch==2:
        n=int(input("enter first number:"))
        o=int(input("enter second number:"))
        print(substract(n,o))
    elif ch==3:
        m=int(input("enter first number:"))
        k=int(input("enter second number:"))
        print(multiply(m,k))
    elif ch==4:
        d=int(input("enter numerator:"))
        u=int(input("enter denominator:"))
        print(divide(d,u))
    else:
        print("enter valid choice")     
      
def add(a,b):
    return a+b
def substract(c,d):
    return c-d
def multiply(e,f):
    return e*f
def divide(g,h):
    return g/h

calculator() 
```

```python
class Rectangle:
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def area(self):
        c=self.a*self.b
        print(c)
    def perimeter(self):
        c=self.a+self.b
        print( c )
r=Rectangle(14,45)
r.area()
r.perimeter()
```
