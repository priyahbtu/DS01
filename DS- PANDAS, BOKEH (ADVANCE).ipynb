```python
from bokeh.models import HoverTool, ColumnDataSource, Text
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()
# Assuming 'elements' is a predefined DataFrame
table = elements
table = table.drop_duplicates(subset='name')
x_data = table["group"]
y_data = table["period"]
element_data = table["name"]
atomic_no_data = table["atomic number"]
mass_no_data = table["atomic mass"]
symbol = table["symbol"]
source = ColumnDataSource(data=dict(x=x_data, y=y_data, element=element_data,
                                    atomic_no=atomic_no_data, mass_no=mass_no_data,
                                    symbol=symbol))
periods = ["I", "II", "III", "IV", "V", "VI", "VII", "-"]
groups = [str(i) for i in range(1, 19)]
p = figure(title="Periodic Table", x_axis_location="above", width=800, height=400,
           x_range=groups, y_range=periods,
x_axis_label="Group", y_axis_label="Period")
text = Text(x="x", y="y", text="symbol", text_align="center", text_baseline="middle")
p.add_glyph(source, text)
tooltips = [("Element", "@element"), ("Atomic Number", "@atomic_no"), 
            ("Mass Number", "@mass_no"), ("Period", "@y"), ("Group", "@x")]
hover = HoverTool(tooltips=tooltips, mode="mouse")
p.add_tools(hover)
show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, LabelSet

# Define marker types and their properties
marker_types = ['circle', 'square', 'triangle', 'diamond', 'cross', 'x']
marker_labels = ['Circle', 'Square', 'Triangle', 'Diamond', 'Cross', 'X']
colors = ['red', 'blue', 'green', 'orange', 'purple', 'cyan'] 
                                       # Distinct colors for each marker type
line_colors = ['black', 'black', 'black', 'black', 'black', 'black'] 
                                       # Line colors for markers
sizes = [10, 12, 14, 16, 18, 20] # Different sizes for each marker type

# Create a figure with no toolbar and a title
p = figure(title="Marker Types with Random Points",
           toolbar_location=None, width=600, height=600)

# Hide the grid lines
p.grid.grid_line_alpha = 0

# Create a ColumnDataSource for labels
all_x = []
all_y = []
all_labels = []

# Generate and scatter points for each marker type
for i in range(len(marker_types)):
    # Generate random points
    x = np.random.randint(1, 10, size=10) # Random x values
    y = np.random.randint(1, 10, size=10) # Random y values
    
    # Scatter the points with specified properties
    p.scatter(x, y, marker=marker_types[i], size=sizes[i],
              color=colors[i], line_color=line_colors[i],
              legend_label=marker_labels[i], alpha=0.8)
   
    # Store the points and labels for the LabelSet
    all_x.extend(x)
    all_y.extend(y)
    all_labels.extend([marker_labels[i]] * len(x)) # Repeat the label for each point

# Create a ColumnDataSource for labels
label_source = ColumnDataSource(data=dict(
    x=all_x,
    y=all_y,
    labels=all_labels
))

# Create a LabelSet for the markers
labels = LabelSet(x='x', y='y', text='labels', source=label_source,
                  x_offset=5, y_offset=5, 
                                 # Position the labels slightly to the right and above
                  text_font_size="10pt", # Set font size
                  text_color="black") # Set label color

# Add the labels to the plot
p.add_layout(labels)

# Show the plot
show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, LabelSet, HoverTool

# Define marker types and their properties
marker_types = ['circle', 'square', 'triangle', 'diamond', 'cross', 'x']
marker_labels = ['Circle', 'Square', 'Triangle', 'Diamond', 'Cross', 'X']
colors = ['red', 'blue', 'green', 'orange', 'purple', 'cyan'] 
                               # Distinct colors for each marker type
sizes = [10, 12, 14, 16, 18, 20] # Different sizes for each marker type

# Create a figure with no toolbar and a title
p = figure(title="Marker Types with Random Points",
           toolbar_location=None, width=600, height=600)

# Hide the grid lines
p.grid.grid_line_alpha = 0

# Create a ColumnDataSource for labels and hover tool
all_x = []
all_y = []
all_labels = []
all_marker_types = []
all_line_colors = []

# Generate and scatter points for each marker type
for i in range(len(marker_types)):
    # Generate random points
    x = np.random.randint(1, 10, size=10) # Random x values
    y = np.random.randint(1, 10, size=10) # Random y values
   
    # Vary line color based on position (y value)
    line_color =[f'#{int(255 * (y_val / 10)):02x}00{int(255 * (1 - (y_val / 10))):02x}' 
     for y_val in y]

    # Scatter the points with specified properties
    scatter = p.scatter(x, y, marker=marker_types[i], size=sizes[i],
                        color=colors[i], line_color=line_color,
                        legend_label=marker_labels[i], alpha=0.8)
    # Store the points and labels for the LabelSet and hover tool
    all_x.extend(x)
    all_y.extend(y)
    all_labels.extend([marker_labels[i]] * len(x)) # Repeat the label for each point
    all_marker_types.extend([marker_types[i]] * len(x)) 
                                  # Repeat the marker type for hover
    all_line_colors.extend(line_color) # Store the line colors

# Create a ColumnDataSource for labels and hover tool
source = ColumnDataSource(data=dict(
    x=all_x,
    y=all_y,
    labels=all_labels,
    marker_types=all_marker_types,
    line_colors=all_line_colors
))

# Create a LabelSet for the markers
labels = LabelSet(x='x', y='y', text='labels', source=source,
x_offset=5, y_offset=5, # Position the labels slightly to the right and above
text_font_size="10pt", # Set font size
text_color="black") # Set label color

# Add the labels to the plot
p.add_layout(labels)

# Add hover tool to display marker type
hover = HoverTool()
hover.tooltips = [("Marker Type", "@marker_types")]
p.add_tools(hover)

# Show the plot
show(p)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

diabetes = datasets.load_diabetes()

diabetes_x = diabetes.data
diabetes_y = diabetes.target

diabetes_x_train, diabetes_x_test, diabetes_y_train, diabetes_y_test = train_test_split(
    diabetes_x, diabetes_y, test_size=0.3, random_state=42
)
model = linear_model.LinearRegression()

model.fit(diabetes_x_train, diabetes_y_train)

diabetes_y_predicted = model.predict(diabetes_x_test)

mse = mean_squared_error(diabetes_y_test, diabetes_y_predicted)
print("Mean Squared Error is:", mse)
print("Weights:", model.coef_)
print("Intercept:", model.intercept_)
plt.scatter(diabetes_y_test, diabetes_y_predicted)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted Values")
plt.plot([min(diabetes_y_test), max(diabetes_y_test)], [min(diabetes_y_test), 
                                                        max(diabetes_y_test)],
         color='red', linestyle='--') # Line for perfect prediction
plt.show()
```

```python
import matplotlib.lines as mlines
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_blobs
from sklearn.inspection import DecisionBoundaryDisplay

# Generate synthetic data
n_samples_1 = 1000
n_samples_2 = 100
centers = [[0.0, 0.0], [2.0, 2.0]]
clusters_std = [1.5, 0.5]

# `make_blobs` creates data for classification
X, y = make_blobs(
    n_samples=[n_samples_1, n_samples_2],
    centers=centers,
    cluster_std=clusters_std,
    random_state=42,
)

# Create an SVM classifier without weights
clf = svm.SVC(kernel="linear", C=1.0)
clf.fit(X, y)

# Create a weighted SVM classifier
wclf = svm.SVC(kernel="linear", class_weight={1: 10})
wclf.fit(X, y)

# Plot the data
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, edgecolors="k")

# Plot decision boundaries
ax = plt.gca()
disp = DecisionBoundaryDisplay.from_estimator(
    clf,
    X,
    plot_method="contour",
    colors="k",
    levels=[0],
    alpha=0.5,
    linestyles=["-"],
    ax=ax,
)
wdisp = DecisionBoundaryDisplay.from_estimator(
    wclf,
    X,
    plot_method="contour",
    colors="r",
    levels=[0],
    alpha=0.5,
    linestyles=["-"],
    ax=ax,
)

# Add a legend
plt.legend(
    [
        mlines.Line2D([], [], color="k", label="non weighted"),
        mlines.Line2D([], [], color="r", label="weighted"),
    ],
    ["non weighted", "weighted"],
    loc="upper right",
)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
iris = datasets.load_iris()
X = iris.data
y = iris.target
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
print("Explained variance ratio:", pca.explained_variance_ratio_)
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y,
                      cmap='viridis', 
                      edgecolor='k',
                      s=100)
plt.title('PCA of Iris Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(scatter, label='Classes')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import pandas as pd

# Load Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(iris['data'], columns=iris['feature_names'])

# Perform k-Means clustering
clustering = KMeans(n_clusters=3, random_state=42)
clustering.fit(iris_df)

# Relabel clusters
relabel = np.array([2, 0, 1]).astype(np.int64)[clustering.labels_]

# Define color themes for visualization
color_theme = np.array(['red', 'green', 'blue'])

# Create subplots for visualization, set to A4 size (8.27 x 11.69 inches)
plt.figure(figsize=(6, 5))

# Ground truth classification
plt.subplot(1, 2, 1)
plt.scatter(
    x=iris_df['petal length (cm)'], 
    y=iris_df['petal width (cm)'], 
    c=color_theme[iris.target], 
    s=50
)
plt.title('Ground Truth Classification')

# k-Means classification
plt.subplot(1, 2, 2)
plt.scatter(
    x=iris_df['petal length (cm)'], 
    y=iris_df['petal width (cm)'], 
    c=color_theme[relabel], 
    s=50
)
plt.title('k-Means Classification')

# Show plot
plt.tight_layout()  # Ensure proper spacing between subplots
plt.show()
```
